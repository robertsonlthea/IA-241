# Analyzing COVID-19 Data in Countries with the Least Cases

## Background

On 9 January 2020, WHO announced the beginning of the COVID-19 pandemic with the discovery of a mysterious corona-virus related pneumonia in Wuhan, China. As of the end of 2020, the United States surpassed 20 million infections from SARS-CoV-2, and more than 346,000 deaths. Globally, cases rose to 83,832,334 and 1,824,590 deaths.

## Data Source

The data was downloaded from the [European Centre for Disease Prevention and Control](https://www.ecdc.europa.eu/en/publications-data/download-todays-data-geographic-distribution-covid-19-cases-worldwide)

```python
%matplotlib inline
import pandas
```

## An Overview of the COVID-19 Data

```python
df = pandas.read_excel('s3://robertson-ia241-2021spring/COVID-19-geographic-disbtribution-worldwide-2020-12-14 (1).xls')

df[:100]
```

## Trend in the Number of Cases Per Day From January to December of 2020

```python
sum_cases_per_day = df.groupby('dateRep').sum()['cases']
```

```python
sum_cases_per_day.plot()
```

## Trend in the Number of Deaths Per Day From January to December of 2020

```python
sum_deaths_per_day = df.groupby('dateRep').sum()['deaths']
```

```python
sum_deaths_per_day.plot()
```

### Explanation of Calculations

Calculations were used to assess the total sum of cases and deaths per day in these two line charts by using df.groupby and grouping the x axis by the 'dateRep.' 

The .sum() was used to calculate total cases or deaths in each chart which is represented by the line on the chart.

### Interpretation

There is an exponential growth in both cases and deaths from January of 2020 to December of 2020.

## Countries with the Smallest Amount of COVID-19 Cases

```python
sum_cases_per_country = df.groupby('countriesAndTerritories').sum()['cases'] 
```

```python
sum_cases_per_country.nsmallest(10).plot.bar()
```

### Explanation and Calculations

Calculations in this graphic were used to compare the total number of cases in the top 10 countries with the least amount of COVID-19 cases using df.grouby ('countriesAndTerritories') with .sum() of 'cases'

### Implications

This bar chart shows that some countries were barely affected by the COVID-19 pandemic either due to strict regulations and quarantines or because of the countries geography
